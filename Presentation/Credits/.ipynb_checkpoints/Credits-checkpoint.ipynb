## Предобработка данных

```python
X = df.drop('A16', 1)
y = df['A16']
```

```python
numeric_cols = ['A2', 'A3', 'A8', 'A11', 'A14', 'A15']    
categorical_cols = list(set(X.columns.values.tolist()) - set(numeric_cols))    
```

```python
def calculate_means(numeric_data):
    means = np.zeros(numeric_data.shape[1])
    for j in range(numeric_data.shape[1]):
        to_sum = numeric_data.iloc[:,j]
        indices = np.nonzero(~numeric_data.iloc[:,j].isnull())[0]
        correction = np.amax(to_sum[indices])
        to_sum /= correction
        for i in indices:
            means[j] += to_sum[i]
        means[j] /= indices.size
        means[j] *= correction
    return pd.Series(means, numeric_data.columns)
```

### Обработка пропущенных значений и преобразование вещественных признаков

```python
#Тут надо бы гистаграмму
```

```python
X['A2'] = X['A2'].apply(lambda x: np.NaN if x =='?' else x)
X['A14'] = X['A14'].apply(lambda x: np.NaN if x =='?' else x)

X_num_float = X[numeric_cols].astype(float)
nan_id = np.argwhere(np.isnan(X_num_float['A14'].values))

X_real_zeros = X_num_float.fillna(0.0)
X_real_mean = X_num_float.fillna(calculate_means(X_num_float))

X_cat = X[categorical_cols].astype(str)

#Преобразование ответов
y_num = y.apply(lambda x: 1 if x=='+' else -1)
```

### Преобразование категориальных признаков

```python
encoder = DV(sparse = False)
X_cat_oh = encoder.fit_transform(X_cat.T.to_dict().values())
```

```python
(X_train_real_zeros, 
 X_test_real_zeros, 
 y_train, y_test) = train_test_split(X_real_zeros, y_num, 
                                     test_size=0.2, 
                                     random_state=0)
(X_train_real_mean, 
 X_test_real_mean) = train_test_split(X_real_mean, 
                                      test_size=0.2, 
                                      random_state=0)
(X_train_cat_oh,
 X_test_cat_oh) = train_test_split(X_cat_oh, 
                                   test_size=0.2, 
                                   random_state=0)
```

```python
def plot_scores(optimizer):
    scores = [[item[0]['C'], 
               item[1], 
               (np.sum((item[2]-item[1])**2)/(item[2].size-1))**0.5] for item in optimizer.grid_scores_]
    scores = np.array(scores)
    plt.semilogx(scores[:,0], scores[:,1])
    plt.fill_between(scores[:,0], scores[:,1]-scores[:,2], 
                                  scores[:,1]+scores[:,2], alpha=0.3)
    plt.show()
```

```python
def plot_roc(fpr, tpr, name):
    plt.plot(fpr, tpr, label=name)
    plt.plot([0, 1], [0, 1], '--', color = 'grey', label = 'random')
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve')
    plt.legend(loc = "lower right")
```

```python
X_test_zeros = np.hstack((X_test_real_zeros, X_test_cat_oh))
X_test_mean = np.hstack((X_test_real_mean, X_test_cat_oh))
```

## Собственный

```python
X_train_zeros = np.hstack((X_train_real_zeros, X_train_cat_oh))
X_train_mean = np.hstack((X_train_real_mean, X_train_cat_oh))
```

```python
from sklearn.preprocessing import StandardScaler

# place your code here
scaler = StandardScaler()

X_train_real_zeros_scaled = scaler.fit_transform(X_train_real_zeros)
X_test_real_zeros_scaled = scaler.fit_transform(X_test_real_zeros)
```

## Готовый солвер

```python
param_grid = {'C': [0.01, 0.05, 0.1, 0.5, 1, 5, 10], 'solver':['liblinear','newton-cg']}

cv = StratifiedKFold(y_train, n_folds=5, shuffle=True)

estimator = LR('l2')
optimizer = GridSearchCV(estimator,param_grid,scoring = 'roc_auc',cv=cv)

optimizer.fit(X_train_zeros,y_train)
plot_scores(optimizer)

fpr, tpr, _ = roc_curve(y_test, optimizer.predict_proba(X_test_zeros)[:,1])

plot_roc(fpr, tpr, 'zeros')

#auc_zeros = roc_auc_score(y_test, optimizer.predict(X_test_zeros))
#auc_zeros
```
